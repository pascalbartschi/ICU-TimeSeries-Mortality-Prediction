```python
import umap
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os

%config InlineBackend.figure_format = 'retina'
%load_ext autoreload
%autoreload 2
```

```python
# Load the data
train = pd.read_csv('../../data/embedding_df_Q42_a.csv') 
validation = pd.read_csv('../../data/embedding_df_Q42_b.csv') 
test = pd.read_csv('../../data/embedding_df_Q42_c.csv') 
```

```python
# Create UMAP model and fit-transform the data
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(train.drop('label', axis=1))

# Create scatter plot
plt.figure(figsize=(10, 8))
scatter = plt.scatter(embedding[:, 0], embedding[:, 1], c=train['label'], cmap='viridis')
plt.colorbar(scatter, label='Label')
plt.title('UMAP visualization of embeddings')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```
